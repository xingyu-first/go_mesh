```python
import numpy as np
import os

def unit_mesh(path, pathto):
    print('{} start'.format(path))
    with open(path) as f:
        counter = 0
        n = 0
        m = 0
        v = []
        fa = []
        for line in f:
            if counter == 0:
                if line.split()[0] != 'OFF':
                    print('This is not an OFF file')
                    break
            elif counter == 1:
                (n, m, mm) = map(int, line.split())
            elif counter < n + 2:
                (x, y, z) = map(float, line.split())
                v.append([x, y, z])
            elif counter < n + m + 2:
                fa.append(line)
                if line[0] != '3':
                    print('not triangle')
                    return
            counter += 1
        v = np.array(v)
        maxv = v.max(axis=0)
        minv = v.min(axis=0)
        scale = max(maxv - minv)
        center = (maxv + minv) / 2.0
        #move to center
        v -= center
        #scale to unit box
        v /= scale
    with open(pathto, 'w') as f:
        f.write('OFF\n')
        f.write('{} {} 0\n'.format(len(v), len(fa)))
        for i in v:
            f.write('{} {} {}\n'.format(i[0], i[1], i[2]))
        for line in fa:
            f.write(line)
    print('{} finished'.format(pathto))
    
path = '/home/first/code/data/ModelNet10/ModelNet10'
pathto = '/home/first/code/data/ModelNet10/unitcenterdata'

list = []
list = [x for x in os.listdir(path)]

for i in list:
    aim = path + '/' + i
    aimto = pathto + '/' + i
    if os.path.isdir(aim):
        list.extend([i + '/' + x for x in os.listdir(aim)])
        if not os.path.exists(aimto):
            os.makedirs(aimto)
    else:
        if aim.split('.')[-1].upper() == "OFF":
            unit_mesh(aim, aimto)
```
