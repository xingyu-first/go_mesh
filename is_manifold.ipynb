```python
def read_obj(path):
    vertices = []
    faces = []
    with open(path, 'r') as f:
        for line in f:
            if line[0] == 'v':
                s = line.split()
                vertices.append([float(s[1]), float(s[2]), float(s[3])])
            if line[0] == 'f':
                s = line.split()
                if len(s) > 4:
                    print("not triangle")
                faces.append([int(s[1]), int(s[2]), int(s[3])])
    son = []
    last = []
    pre = []
    twoedge = []
    for i in range(len(vertices) + 1):
        last.append(-1)
    for f in faces:
        for i in range(3):
            x = f[i]
            y = f[(i + 1) % 3]
            
            ok = True
            p = last[x]
            while p > 0:
                yy = son[p]
                p = pre[p]
                if yy == y:
                    twoedge[(p // 2)] += 1
                    ok = False
            if ok:
                s = len(son)
                son.append(y)
                pre.append(last[x])
                last[x] = s
                twoedge.append(0)
                s += 1
                son.append(x)
                pre.append(last[y])
                last[y] = s
    for i in twoedge:
        if i != 2:
            print("border",i)
def read_off(path):
    vertices = []
    faces = []
    counter = 0
    n = 0
    m = 0
    with open(path, 'r') as f:
        for line in f:
            if counter == 0:
                pass
            elif counter == 1:
                s = line.split()
                n = int(s[0])
                m = int(s[1])
            elif counter <= 1 + n:
                s = line.split()
                vertices.append([float(s[0]), float(s[1]), float(s[2])])
            elif counter <= 1 + n + m:
                s = line.split()
                if len(s) > 4:
                    print("not triangle")
                faces.append([int(s[1]), int(s[2]), int(s[3])])
    son = []
    last = []
    pre = []
    twoedge = []
    for i in range(len(vertices) + 1):
        last.append(-1)
    for f in faces:
        for i in range(3):
            x = f[i]
            y = f[(i + 1) % 3]
            
            ok = True
            p = last[x]
            while p > 0:
                yy = son[p]
                p = pre[p]
                if yy == y:
                    twoedge[(p // 2)] += 1
                    ok = False
            if ok:
                s = len(son)
                son.append(y)
                pre.append(last[x])
                last[x] = s
                twoedge.append(0)
                s += 1
                son.append(x)
                pre.append(last[y])
                last[y] = s
    for i in twoedge:
        if i != 2:
            print("border",i)  
    print('sa')
            

p = 'test.off'
read_off(p)
```
